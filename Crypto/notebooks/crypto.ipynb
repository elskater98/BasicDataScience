### 1. Read Data

```python
import sys, os, glob
import pandas as pd
```

```python
REL_PATH = "../data/raw/*.csv"
files = list()

for file in glob.glob(REL_PATH):
    print(file)
    df = pd.read_csv(file, index_col=None, header=0)
    files.append(df)
```

```python
data = pd.concat(files, axis=0, ignore_index=True)
```
